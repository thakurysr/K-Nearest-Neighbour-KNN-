# K Nearest Neighbour

```python
array=df.values
```

```python
x = array[:,:-1]
y = array[:,-1]
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

# Model Building via Training and Testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
# Here we are write a for loop to find the all possible best k values.

score=[]
for i in range(3,30,2):
    model=KNeighborsClassifier(i)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    score.append(accuracy_score(y_test,pred))
```

```python
score
```

```python
max(score)
```

```python
plt.figure(figsize=(10,4))
plt.bar(range(3,30,2),score)
plt.xticks(range(3,30,2))
plt.xlabel("K-Values",size=20)
plt.ylabel("Accuracy",size=20)
plt.ylim(0.7,1)
plt.style.use("seaborn-darkgrid")
plt.scatter(range(3,30,2),score,s=100,color="Black")
plt.show()
```

# Cross Validation

### Here we are taking k value manually.

### K=3

```python
knn=KNeighborsClassifier(n_neighbors=3)
score=cross_val_score(knn,x_train,y_train,cv=5)
score.mean()
```

### K=5

```python
knn=KNeighborsClassifier(n_neighbors=5)
score=cross_val_score(knn,x_train,y_train,cv=5)
score.mean()
```

### K=7

```python
knn=KNeighborsClassifier(n_neighbors=7)
score=cross_val_score(knn,x_train,y_train,cv=5)
score.mean()
```

### K=9

```python
knn=KNeighborsClassifier(n_neighbors=9)
score=cross_val_score(knn,x_train,y_train,cv=5)
score.mean()
```

### K=11

```python
knn=KNeighborsClassifier(n_neighbors=11)
score=cross_val_score(knn,x_train,y_train,cv=5)
score.mean()
```

```python
# for loop for best possible k values

scores=[]

for i in range(3,30,2):
    model=KNeighborsClassifier(i)
    score=cross_val_score(model,x_train,y_train,cv=5)
    scores.append(score.mean())
```

```python
scores
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.bar(range(3,30,2),scores)
plt.xticks(range(3,30,2))
plt.xlabel("K-value",size=20)
plt.ylabel("Accuracy",size=20)
plt.ylim(0.5,1)
plt.scatter(range(3,30,2),scores,s=100,color="Black")
plt.show()
```

### We can see that our best k value is K=3

```python
# plot for k value
plt.plot(range(3,30,2),scores, marker='o')
plt.title('Accuracy of Model for Varying Values of K')
plt.xlabel("Values of K")
plt.ylabel("Mean Accuracy Score")
plt.xticks(range(3,30,2))
plt.rcParams['figure.figsize'] = (5,5) 
plt.show()
```

# Final Model

```python
final_model=KNeighborsClassifier(3)
final_model.fit(x_train,y_train)
pred=final_model.predict(x_test)
accuracy_score(y_test,pred)
```
